```python
import somoclu
import numpy as np
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt
```

```python
data,labels = make_blobs(n_samples=10000,n_features=2,cluster_std=0.5,random_state=42)
```

```python
z_label = (labels == 0)
o_label = (labels == 1)
t_label = (labels == 2)
```

```python
plt.scatter(data[z_label][:,0],data[z_label][:,1],c='red')
plt.scatter(data[o_label][:,0],data[o_label][:,1],c='green')
plt.scatter(data[t_label][:,0],data[t_label][:,1],c='blue')
plt.show()
```

```python
data = data.astype('float32')
```

```python
data.dtype
```

```python
n_rows, n_columns = 100,100

som = somoclu.Somoclu(n_rows,n_columns,compactsupport=True)
%timeit som.train(data)
```

```python
som.view_component_planes()
```
